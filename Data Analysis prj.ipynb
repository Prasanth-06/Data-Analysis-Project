```python
import pandas as pd
data=pd.read_csv("supermarket.csv")
data
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

### The product sales per hour can be anaylsed with a lineplot.

```python
data["Time"]= pd.to_datetime(data["Time"])
data["Hour"]= (data["Time"]).dt.hour
plt.figure(figsize=(14,6)) 
SalesTime = sns.lineplot(x="Hour", y ="Quantity", data = data).set_title("product sales per Hour")
```

## OBSERVATION:

We can see that the supermarket makes most of its sells in 14:00 hrs local time.

### ANALYSIS: The analysis of different types of customers is done in a count plot.

```python
plt.figure(figsize= (14,6))
ax = sns.countplot(x = "Customer type", data = data, palette = "rocket_r")
ax.set_title("Type of customers", fontsize = 25)
ax.set_xlabel("Customer type", fontsize = 16)
ax.set_ylabel("Customer Count", fontsize = 16)
plt.show()
```

### OBSERVATION:

The type of customer is analysed in count plot.Here we observe the occurences of a member and normal customer.

### ANALYSIS:  We can analyse the average rating of branch A and C with a box plot.

```python
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =data, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Rating distribution", fontsize = 16)
plt.show()
```

## OBSERVATION:

We can see that the average rating of branch A and C is more than seven and branch B is less than 7.

### ANALYSIS:  Using the boxenplot, we can analyse the average sales of different products.

```python
plt.figure(figsize = (14,6))
ax = sns.boxenplot(y = "Product line", x = "Rating", data = data)
ax.set_title("Average sales of different products", fontsize = 25)
ax.set_xlabel("Rating", fontsize = 16)
ax.set_ylabel("Product line", fontsize = 16)
plt.show()
```

### OBSERVATION:

Here we can see that the average sales of different lines of products. Health and beauty making the highest sales whereas Fashon accessories making the lowest sales.

### ANALYSIS: The sales count can be done using a count plot.

```python
plt.figure(figsize=(14,6))
ax = sns.countplot(y='Product line', data=data, order = data['Product line'].value_counts().index)
ax.set_title(label = "Sales count of products", fontsize = 25)
ax.set_xlabel(xlabel = "Sales count", fontsize = 16)
ax.set_ylabel(ylabel= "Product Line", fontsize = 16)
```

### OBSERVATION:

We can see the top sold products like the health and beauty,sports and travel,food and bevarages etc form the above figure.

### ANALYSIS: The total sales of products can be analysed with a scatter plot.

```python
plt.figure(figsize = (14,6))
ax= sns.stripplot(y= "Product line", x = "Total", hue = "Gender", data = data)
ax.set_title(label = "Product sales on the basis of gender")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Product Line")
plt.show()
```

### OBSERVATION:

Product sales on the basis of gender is done using the scatter plot where we have the gender like male and female.Here we observe the relationship between the gender with respect to product line and total sale products

### ANALYSIS: The product line rating for all the products is given by a violin plot.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize = (14,7))
data = pd.read_csv("supermarket.csv")
sns.violinplot(x=data["Product line"],y=data["Rating"])
plt.show()
```

### OBSERVATION:
Product line rating is done with a violin plot.Here we observe the distribution of all the products.

### ANALYSIS: The analysis of cities in done in a donut

```python
import matplotlib.pyplot as plt

labels = 'yangon', 'Naypyitaw', 'manglay'
sizes = [15, 30, 45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
        
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```

### OBSERVATION:
The analysis of cities is done in a donut.Here we observe the proportions of each city with the size of each piece.

### ANALYSIS: The analysis between product line and unit price is analysed with a piechart.

```python

import matplotlib.pyplot as plt

Productline             = ['Health and beauty','Electronic accessories','Home and lifestyle','Sports and travel','Food and beverages','Fashion accessories']

Unitprice   = [74.69,15.28,46.33,86.31,88.36,33.52]
elpo=[0,0,0,0,5,0]
 
figureObject, axesObject = plt.subplots()

axesObject.pie(Unitprice,

        labels=Productline,

        autopct='%1.2f',

        startangle=90)

axesObject.axis('equal')

plt.show()
```

### OBSERVATION:
The analysis of product line and unit price is done in a pie chart.Here we observe the percentage of products and the unit price.
